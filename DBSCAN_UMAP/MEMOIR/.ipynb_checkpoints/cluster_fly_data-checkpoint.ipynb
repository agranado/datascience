```python
import requests
import os
import pandas as pd
import umap
import numpy as np
#from ggplot import *
from dfply import *
#from plotnine import *

#for scaling the dataset
from sklearn.preprocessing import StandardScaler
# for PCA
from sklearn.decomposition import PCA
#for plotting
import matplotlib.pyplot as plt
# for dbscan
from sklearn.cluster import DBSCAN

import sklearn.cluster as cluster
```

```python
## this data frame was pre-processed in R to include the clone_id column 
# clust_label is an old column from first pass UMAP 

fly_data = pd.read_csv("fly_data_clone_id.txt")
fly_data = pd.read_csv("full_matrix_11132019.csv")
fly_data = pd.read.csv("full_matrix_11132019_filterRows.csv")

fly_data.head()
```

```python
# let's filter for cells with actual clone ID

#fly_filtered = fly_data >> mask(X.clust_label >=0 , X.clone_id >0)
# filter only for cells that belong to a clone (this is optional)
fly_filtered = fly_data >> mask(X.cell_size>1)

# Fig 2 
# Cluster all cells even if they dont have a barcode 
fly_filtered = fly_data >>  mask(X.cell_size>1)


#where are the genes
cols = fly_filtered.columns
# old matrix : features = cols[2:10]
features = cols[5:13]

fly_filtered.head()
```

```python
features

#
x = fly_filtered.loc[:,features].values


fly_filtered.head()
fly_filtered.shape


# since the measurement of gene expression is a sum over all pixels, we should normalize by the cell size. 
# Also this brings the numbers closer together 
x = fly_filtered.loc[:,features].div(fly_filtered.iloc[:,4], axis=0)
x
```

```python
x =np.log(x+1)
```

```python
x
```

```python
x_scaled = StandardScaler().fit_transform(x)
pca = PCA(n_components =6)
lowd_x = pca.fit_transform(x_scaled)

pca.explained_variance_ratio_.cumsum()
```

```python
# UMAP 1 for visualization
standard_embedding = umap.UMAP(random_state = 42).fit_transform(lowd_x)
```

```python
# UMAP 2 this is a UMAP with different parameters, not for visualization but for clustering
clusterable_embedding = umap.UMAP(
    n_neighbors=25,
    min_dist=0.0,
    n_components=3,
    random_state=42,
).fit_transform(lowd_x)

clusterable_embedding

# # # CLUSTERING
dbscan_labels = DBSCAN(eps = 0.38).fit_predict(clusterable_embedding)

#How many clusters
np.unique(dbscan_labels)

# Size distribution of clusters
pd.value_counts(dbscan_labels)
```

```python

clustered =(dbscan_labels >=0)

pd.value_counts(dbscan_labels[clustered])

#if we are happy with the results we can save to data frame 

#save the cluster labels
fly_filtered.loc[:,'clust_label'] = dbscan_labels
fly_filtered.loc[:,'umap_1'] = standard_embedding[:,0]
fly_filtered.loc[:,'umap_2'] = standard_embedding[:,1]

```

```python
# UMAP PLOT
# Fitler for cluster -1
clustered =(dbscan_labels >=0)


point_size = 6
f = plt.figure()
plt.scatter(standard_embedding[~clustered,0],standard_embedding[~clustered,1],c = (0.5,0.5,0.5),s = point_size,alpha =0.5)


plt.scatter(standard_embedding[clustered, 0],
            standard_embedding[clustered, 1],
            c=dbscan_labels[clustered],
            s=point_size, label = dbscan_labels[clustered],
            cmap='Spectral');
plt.ylabel("UMAP 1")
plt.xlabel("UMAP 2")
plt.show()
```
